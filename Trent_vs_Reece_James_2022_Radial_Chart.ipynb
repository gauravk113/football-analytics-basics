```python
from mplsoccer import Radar, FontManager
import matplotlib.pyplot as plt
```

```python
# parameter names of the statistics we want to show
params = ["npxG", "Non-Penalty Goals", "xA", "Assists",  "Key Passes","Through Balls",
          "Progressive Passes", "Shot-Creating Actions", "Goal-Creating Actions",
          "Dribbles Success", "Pressure Regains", "Touches In Box", "Crosses", "Tackles Completed", "Tackles Won Possession", "Dribbled Past", "Blocks", "Interceptions", "Aerial Duels Won", "Pass Completion", "Goals/Shot"]

# The lower and upper boundaries for the statistics
low =  [0.05, 0.0, 0.1, 0.2, 0.7, 0.10, 3, 0.8, 0.2, 0.5, 1, 0.8, 1.5, 1, 0.4, 0.4, 0.7, 0.6, 0.25, 50, 0.01]
high = [0.2, 0.43, 0.5, 0.51, 3.2, 1, 11, 5.5, 1, 2.5, 5.5, 4.35, 4.3, 2, 1.8, 2, 2, 3, 1.5, 100, 0.35]
```

```python
radar = Radar(params, low, high,
              # whether to round any of the labels to integers instead of decimal places
              round_int=[False]*21,
              num_rings=3,  # the number of concentric circles (excluding center circle)
              # if the ring_width is more than the center_circle_radius then
              # the center circle radius will be wider than the width of the concentric circles
              ring_width=1, center_circle_radius=1)
```

```python
trent_values =  [0.09, 0.06, 0.35, 0.38, 2.81, 0.63, 8.77, 4.73, 0.54, 0.82, 3.06, 2.81, 4.01, 1.39, 0.76, 1.14, 1.23, 2.40, 0.57, 74, 0.04]
reece_values = [0.13, 0.24, 0.28, 0.43, 2.27, 0.19, 5.80, 3.81, 0.53, 1.64, 4.49, 3.62, 3.96, 1.88, 1.11, 0.82, 1.40, 1.01, 1.06, 86.5, 0.14 ] 
```

```python
fig, ax = radar.setup_axis()  # format axis as a radar
rings_inner = radar.draw_circles(ax=ax, facecolor='#ffb2b2', edgecolor='#fc5f5f')  # draw circles
radar_output = radar.draw_radar(trent_values, ax=ax,
                                kwargs_radar={'facecolor': '#aa65b2'},
                                kwargs_rings={'facecolor': '#66d8ba'})  # draw the radar
radar_poly, rings_outer, vertices = radar_output
range_labels = radar.draw_range_labels(ax=ax, fontsize=9,
                                       )  # draw the range labels
param_labels = radar.draw_param_labels(ax=ax, fontsize=8,
                                       )  # draw the param labels
```

```python
# plot radar
fig, ax = radar.setup_axis()
rings_inner = radar.draw_circles(ax=ax, facecolor='grey', edgecolor='pink')
radar_output = radar.draw_radar_compare(trent_values, reece_values, ax=ax,
                                        kwargs_radar={'facecolor': 'red', 'alpha': 0.6},
                                        kwargs_compare={'facecolor': 'blue', 'alpha': 0.6})
                                        
radar_poly, radar_poly2, vertices1, vertices2= radar_output
range_labels = radar.draw_range_labels(ax=ax, fontsize=10)
param_labels = radar.draw_param_labels(ax=ax, fontsize=10)
```
