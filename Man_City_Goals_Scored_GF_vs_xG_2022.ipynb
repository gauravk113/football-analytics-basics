```python
df = pd.read_csv('city_g_vs_xg_2022_final.csv', index_col=0)
```

```python
df
```

```python
opponent = df.index.values.tolist()
amount_data = len(opponent)

x = np.linspace(1, 100, len(opponent))
```

```python
rg_list = df['goals_90'].tolist()
xg_list = df['xg_90'].tolist()
```

```python
rg_list = df['goals_90'].tolist()
xg_list = df['xg_90'].tolist()

rg = np.array(rg_list)
xg = np.array(xg_list)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.plot(x, xg, color='k', lw=2)
plt.title('Manchester City FC Evolution Of Goals Scored (GF) vs Expected Goals (xG) 2021-22 Premier League', fontsize=16, weight='bold')
ax.fill_between(
    x,
    xg,
    rg,
    where=rg >= xg,
    facecolor='#2225e3',
    alpha=0.7,
    interpolate=True,
    label='Where Performance exceeds Expected Goals (xG) ')
ax.fill_between(
    x,
    xg,
    rg,
    where=rg < xg,
    facecolor='#e82c2c',
    alpha=0.7,
    interpolate=True,
    label='Where Performance is below Expected Goals (xG)')
ax.annotate(
    'Black Line xG for 90 minutes',
    xy=(9, 0.135),
    size=8,
    ha="center",
    color='0.3')
plt.ylim((0, 0.15))
plt.xticks(x, opponent, rotation='vertical')
plt.tick_params(axis='both', which='major', labelsize=13)
plt.legend(fontsize='x-large')
ax.set_facecolor('whitesmoke')
plt.grid('off')
plt.tight_layout()
plt.show()
```
